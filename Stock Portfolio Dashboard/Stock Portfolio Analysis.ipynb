## I. Data Extraction and Cleaning

```python
import pandas as pd
import numpy as np
import datetime as dt
import plotly.graph_objs as go
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import random
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
```

```python
## Set dates for data extraction
init_date = dt.datetime(2017,1,1)
end_date = dt.datetime(2019,7,26)
year_end = dt.datetime(2018,12,31)
```

```python
## First Get SP500 data and dates of trading activities
sp500 = pdr.get_data_yahoo('^GSPC',init_date,end_date+dt.timedelta(days = 2)).reset_index()
trading_dates = [date for date in sp500['Date'].unique()]
```

```python
## Produce a hypothetical portfolio of stocks acquired in 2017, intended to be sold during 2019 
## This example includes a mix of  tech and CPG companies (FB,AMZN,SHOP,MSFT,PG,JNJ)

random.seed(10)
ticker_names = ['FB','AMZN','SHOP','MSFT', 'NFLX', 'PG','JNJ','KHC']


def get_portfolio(init_date,year_end,ticker_names, trading_dates):    
    random_number_of_days = [random.randrange(5,100) for _ in range(len(ticker_names))]
    purchase_dates = [dt.datetime.utcfromtimestamp(trading_dates[num_days].tolist()/1e9) for num_days in random_number_of_days]
    quantity = [random.randrange(50, 200, 10) for _ in range(len(ticker_names))]

    unit_cost = []
    for name, purchase_date in zip(ticker_names,purchase_dates):
        value = pdr.get_data_yahoo(name, purchase_date, purchase_date+dt.timedelta(days=2))
        unit_cost.append(value['Adj Close'][0].round(4))
        print(name, purchase_date, value['Adj Close'][0].round(0))
        
    portfolio = pd.DataFrame({'Purchase Date': purchase_dates, 
                          'Ticker': ticker_names,
                          'Quantity': quantity,
                          'Unit Cost': unit_cost})
    
    portfolio['Total Cost'] = portfolio['Quantity']*portfolio['Unit Cost']
    portfolio['Start of Year'] = year_end
    
    return portfolio

portfolio = get_portfolio(init_date,year_end,ticker_names,trading_dates)
```

```python
portfolio
```

```python
## Get present-day portfolio adjusted closing prices
## Data from plus/minus 3 days were obtained to avoid having to download too much unncessary data 

def ticker_price_now(portfolio, ticker_names, end_date, year_end):
    price_now = []
    price_year_end = []
    
    for name in ticker_names:
        present_price = pdr.get_data_yahoo(name, end_date - dt.timedelta(days=3), end_date + dt.timedelta(days=3)).reset_index()
        present_price = present_price.loc[present_price['Date']==end_date,'Adj Close'].iat[0].round(4)
        price_now.append(present_price)
        year_end_price = pdr.get_data_yahoo(name, year_end - dt.timedelta(days=3), year_end + dt.timedelta(days=3)).reset_index()
        year_end_price = year_end_price.loc[year_end_price['Date']==year_end,'Adj Close'].iat[0].round(4)
        price_year_end.append(year_end_price)
        
    portfolio['Present Date'] = end_date
    portfolio['Year-end Price'] = price_year_end
    portfolio['Present Price'] = price_now
    
    return portfolio

data = ticker_price_now(portfolio, ticker_names, end_date, year_end) 
```

```python
## Get all SP500 values relevant: initial price (price on the date of purchase) and present price

def get_sp500(sp500, data, end_date, year_end):
    date_acquired = data['Purchase Date'].unique()
    date_acquired = [dt.datetime.utcfromtimestamp(date.tolist()/1e9) for date in date_acquired]
    filtered = sp500.loc[sp500['Date'].isin(date_acquired)]
    filtered = filtered[['Date','Adj Close']]                         
    data = data.merge(filtered, how = 'left', left_on = 'Purchase Date', right_on = 'Date')
    data.rename(columns = {"Adj Close":"SP500 Initial Price"}, inplace = True)
    data = data.drop('Date', axis=1)
    
    data['SP500 Year-end Price'] = sp500.loc[sp500['Date']==year_end,'Adj Close'].iat[0].round(4)
    data['SP500 Present Price'] = sp500.loc[sp500['Date']==end_date,'Adj Close'].iat[0].round(4)
    
    return data

data = get_sp500(sp500, data, end_date, year_end)
```

We now have a well-organized table of all the data we need to compare our own porfolio to the performance of SP500 index. The next step would be to calculate performance metrics to get a closer look at our porfolio. The dateframe will end up having many columns to include all the metrics needed.

## II. Comparison of Portfolio against SP500

This part of the process is a bit tedious - There are 3 metrics that I would like to examine for this portfolio and each has to be coded somehow manually.

The performances metrics are:

1. Return: the present price minus the initial (purchase) divded by initial (purchase) price
2. Share value: total value of the stock at the moment
3. Gain/Loss: total value of the stock minus the total cost of the stock 


```python
## Calculate return for portfolio vs. SP500

data['Stock Return'] = data['Present Price']/data['Unit Cost'] - 1
data['SP500 Return'] = data['SP500 Present Price']/data['SP500 Initial Price'] - 1

## Calculate absolute share value of stock value porfolio vs. SP500

data['Eq SP500 Shares'] = data['Total Cost']/data['SP500 Initial Price']
data['Stock Share Value'] = data['Quantity'] * data['Present Price']
data['SP500 Share Value'] = data['Eq SP500 Shares'] * data['SP500 Present Price']

## Calcualte gain/loss of stock vs. SP500

data['Stock Gain'] = data['Stock Share Value'] - data['Total Cost']
data['SP500 Gain'] = data['SP500 Share Value'] - data['Total Cost']
```

Additionally, it would be interesting to see how the portfolio grows in value over time compared to if we were to just purchase the SP500 index. To do this, we will need to turn the data into a time-series. We'd need to do one more data extraction using yahoo finance in this case. We will just look into 2018 data for now.

```python
## Get all data relevant to our portfolio
full_portfolio = pdr.get_data_yahoo(ticker_names, dt.datetime(2018,1,1), dt.datetime(2018, 12, 31)).reset_index()
```

```python
## Include only relevant columns
full_portfolio = full_portfolio[['Date','Adj Close']]
## Reduce level of index
full_portfolio.columns = ["_".join(x) for x in full_portfolio.columns.ravel()]
## Set date to be index
full_portfolio.set_index('Date_', inplace = True)
## Rename Columns
full_portfolio.columns = [str(x)[10:] for x in full_portfolio.columns]
## Turn date back into a column
full_portfolio.reset_index(inplace = True)
full_portfolio.rename(columns={'Date_':'Date'}, inplace = True)
## "Melt" the column so it is easier for us to sum up later
full_portfolio = pd.melt(full_portfolio, id_vars = 'Date', var_name = 'Ticker')
## Join the full data to portfolio data containing quantity of stock purchased
full_portfolio = full_portfolio.merge(portfolio[['Ticker', 'Quantity']], on = 'Ticker')
## Calculate value by date
full_portfolio['Stock Value'] = full_portfolio['value']*full_portfolio['Quantity']
```

```python
## Now aggregate at the daily level to get the overall portfolio value by day
daily_summary = full_portfolio.groupby('Date').agg({'Stock Value':'sum'}).reset_index()
## Join to SP500 data
daily_summary = daily_summary.merge(sp500[['Date','Adj Close']], on = 'Date')
## Multiply SP500 by the appropriate number of shares
sp500_eq = sum(data['Eq SP500 Shares'])
daily_summary['SP500 Value'] = sp500_eq * daily_summary['Adj Close']
## Add base cost to purchase the stocks in the fist place
total_cost = sum(data['Total Cost'])
daily_summary['Base Cost'] = total_cost
## Drop irrelevant Columns
daily_summary.drop(columns = 'Adj Close', inplace = True)
```

## III. Data Visualization

The next step of the analysis is to visualize the performance of this randomly generated portfolio against the SP500 index. SP500 essentially serves as a benchmark - often times it is actually much less risky and still a stable way of investing if one just pick to invest in an index fund. So it is important that if we were to choose to create our own portfolio, we should at least beat the index funds.

```python
fig = go.Figure()


fig.add_trace(
    go.Bar(y = data['Stock Return'], x = data['Ticker'], 
              name = 'Stock Return', width = 0.3,
              marker_color='#1f77b4', marker_line_color='#1f77b4'
              )
)

fig.add_trace(
    go.Bar(y = data['SP500 Return'], x = data['Ticker'], 
               name = 'SP500 Return',width = 0.3,
              marker_color='#ff7f0e', marker_line_color='#ff7f0e'
          )
               
)

fig.update_layout(
    title_text="Portfolio vs. SP500 Return", title_x = 0.5
)

fig.show("png",height = 800, width = 1200)
```

Here we can see that almost all the stocks in the portfolio have outperformed the index fund - except KHC (Kraft-Heinz). We also see that the index fund has almost the same return regardless the timing of purchase within 2017. This is unsurprising given that index funds are usually steady in growth but at a much lower rate than individual stocks. The same observation is expected from plotting the gain/loss, but now we will see the impact of the number of shares, as in the next plot.

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(y = data['Stock Gain'], x = data['Ticker'], 
              name = 'Stock Gain/Loss', width = 0.3,
              marker_color='#1f77b4', marker_line_color='#1f77b4'
              )
)

fig.add_trace(
    go.Bar(y = data['SP500 Gain'], x = data['Ticker'], 
               name = 'SP500 Gain/Loss',width = 0.3,
              marker_color='#ff7f0e', marker_line_color='#ff7f0e'
          )
               
)

fig.update_layout(
    title_text="Portfolio vs. SP500 Gain/Loss", title_x = 0.5
)

fig.show("png",height = 800, width = 1200)
```

So here we see that, in absolute values, we have made the most money from Amazon, Shopify, and Netflex - all based on a randomly selected amount of shares to buy! Unsurprisingly, the growth of large CPG company's stock values are not as impressive as tech companies. In fact, in the case of J&J and Kraft-Heinz, things would've been better if the funds were used to purchase idex funds. 

Next I am curious to see the growth of this fictional portfolio over time (2018). We will use the time-series summary data for this

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(y = daily_summary['Stock Value'], x = daily_summary['Date'], 
              name = 'Portfolio Value',               
              line = dict(color='royalblue', width=2)
              )
)

fig.add_trace(
    go.Scatter(y = daily_summary['SP500 Value'], x = daily_summary['Date'], 
               name = 'SP500 Value',
              line = dict(color='firebrick', width=2)
          )
               
)

fig.add_trace(
    go.Scatter(y = daily_summary['Base Cost'], x = daily_summary['Date'], 
               name = 'Base Cost',
              line = dict(color='gray', width=2)
          )
               
)

fig.update_layout(
    title_text="Portfolio vs. SP500 Value in 2018", title_x = 0.5
)

fig.show("png",height = 800, width = 1200)
```

There are a few things to note from this graph:

1. Despite having a lower overall value, an hypothetical all-SP500 investment still allows for positive gains. Of course, in reality, we would also have to take discount rate into consideration
2. Fluctuations in the SP500 and Portfolio Values seem to follow similar trends
3. Overall, the portfolio still performs better than the benchmark index fund - this is good news!
